```python
import requests

# API endpoint (example)
url = "https://api.nasa.gov/insight_weather/?api_key=DEMO_KEY&feedtype=json&ver=1.0"

params = {
    "key": 'HqIO6LmR7CotOVxxGRIBl7laceIdSYs5X71hxTWr',
    "query": "sample_query",  # Exemplo de parâmetro
}

# Make the GET request
response = requests.get(url, params=params)

# Parse the JSON response
if response.status_code == 200:
    json_data = response.json()  # Automatically parses JSON
    print(json_data)     # Access 'data'
else:
    print(f"Error: {response.status_code}")
```

```python
import pandas as pd

# Extract data from the JSON
items = json_data["data"]

# Convert to DataFrame
df = pd.DataFrame(items)

# Preview the data
print(df.head())

# Save to CSV
df.to_csv("output.csv", index=False)
```

```python
import requests

# Endpoint da API
url = "https://api.nasa.gov/insight_weather/?api_key=DEMO_KEY&feedtype=json&ver=1.0"

# Parâmetros da requisição
params = {
    "key": 'HqIO6LmR7CotOVxxGRIBl7laceIdSYs5X71hxTWr',
    "query": "sample_query",  # Exemplo de parâmetro
}

# Requisição GET
response = requests.get(url, params=params)

# Verifique o status e extraia os dados
if response.status_code == 200:
    data = response.json()  # Converte a resposta em formato JSON
    print(data)
else:
    print(f"Erro: {response.status_code}, {response.text}")
```

```python
data
```

```python
import pandas as pd

# Transformar os dados em um DataFrame
df = pd.DataFrame(data)  # Ajuste conforme o formato da API
print(df.head())

# Salvar em CSV
df.to_csv("dataset.csv", index=False)
```
